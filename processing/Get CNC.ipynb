```python
import pandas as pd
import numpy as np


def run(st1,st2):
    cols = ['corpus','doc_id','sent_id','eg_id','index','text','text_w_pairs','seq_label','pair_label','context','num_sents']

    # merge
    data = st1[['index','text','label']].merge(
        st2[['doc_id','eg_id','text','text_w_pairs']], 
        how='outer', left_on='index', right_on='doc_id'
    )
    data['corpus']='cnc'
    data['text'] = data['text_y'].fillna(data['text_x'])
    data = data.drop(columns=['doc_id','text_x','text_y'])
    data = data.rename(columns={'index':'doc_id','label':'pair_label'})
    data['seq_label'] = data['pair_label']
    data['sent_id'] = 0 # all individual sentences
    data['context'] = ''
    data['num_sents'] = 1
    data['eg_id'] = data['eg_id'].fillna(0)
    data['eg_id'] = data['eg_id']
    data['seq_label'] = data.groupby(['corpus','doc_id','sent_id'])['seq_label'].transform('max')
    data['text_w_pairs'] = data['text_w_pairs'].fillna('')
    data['index'] = data['corpus']+'_'+data['doc_id'].astype(str)+'_'+data['sent_id'].astype(str)+'_'+data['eg_id'].astype(int).astype(str)
    data = data[cols]
    return data
```

```python
st1_file = r"D:\61 Challenges\2022_CASE_\CodaLab\03. Official Datasets\Subtask 1\all_train_subtask1.csv"
st2_file = r"D:\61 Challenges\2022_CASE_\WebAnno\reviewing_annotations\Subtask2\07. Round5\curation\reviewed_all_train.csv"

st1 = pd.read_csv(st1_file)
st2 = pd.read_csv(st2_file)
train_df = run(st1,st2)
train_df
```

```python
st1_file = r"D:\61 Challenges\2022_CASE_\CodaLab\03. Official Datasets\Subtask 1\test_subtask1.csv"
st2_file = r"D:\61 Challenges\2022_CASE_\WebAnno\reviewing_annotations\Subtask2\07. Round5\curation\reviewed_all_test.csv"

st1 = pd.read_csv(st1_file)
st2 = pd.read_csv(st2_file)
test_df = run(st1,st2)
test_df
```

```python
data = pd.concat([train_df, test_df], axis=0)
data
```

```python
data.dropna().to_csv('cleaned/cnc.csv', index=False,encoding='utf-8-sig')
```
